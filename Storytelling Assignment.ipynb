```python
en_gen = pd.read_excel('EntriesGender.xlsx')
en_gen.head()
```

```python
medals = pd.read_excel('Medals.xlsx')
medals.head()
```

```python
athletes.info()
coaches.info()
en_gen.info()
medals.info()
teams.info()
```

```python
a_pivot = pd.pivot_table(athletes, values = ['Name', 'Discipline'], index = 'NOC', aggfunc = {'Name':lambda x: len(x.unique()), 'Discipline':lambda x: len(x.unique())})
a_pivot.head()
```

```python
medals.rename(columns = {'Team/NOC':'NOC'}, inplace = True)
medals.head()
```

```python
c_info = medals.join(a_pivot,on = 'NOC',how = 'outer')
c_info.head()
```

```python
c_info.fillna(0, inplace = True)
c_info.info()
```

```python
sns.scatterplot(data = c_info, x = "Discipline",y = "Name", hue = 'Total', size = 'Gold')
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left', borderaxespad=0)
plt.title('Olympics 2021 - Larger Team, More Discplines, More Medals, More Gold')
plt.xlabel('Number of Sports per Country')
plt.ylabel('Number of Athletes per Country')
```

```python
c_info['p_gold'] = round(c_info['Gold']/c_info['Total'],1)
sns.boxplot(data = c_info, x = "p_gold",y = "Discipline")
plt.title('More Active Countries have 30% Gold Medals')
plt.xlabel('Percent of Gold')
plt.ylabel('Number of Sports per Country')
```

Plot Ideas

Scatter Plot:
x = number of sports
y = number of athletes
color = number of medals

Bar Plots:
Per country, count of men medals vs women medals

Questions:

```python
low_medal_countries = c_info[c_info['Total']<= 10].NOC
plt.figure(figsize = (15,5))
sns.countplot(data = athletes[athletes.NOC.isin(low_medal_countries)].sort_values(by = 'Discipline', ascending = True), x = 'Discipline')
plt.xticks(rotation = 90)
plt.title('Sports that Countries with Less than 10 Medals Participated in Olympics 2021')
```

```python
no_medal_countries = c_info[c_info['Total']== 0].NOC
plt.figure(figsize = (15,5))
#sns.countplot(data = athletes[athletes.NOC.isin(no_medal_countries)].sort_values(by = 'Discipline', ascending = True), x = 'Discipline')
sns.countplot(data = athletes[athletes.NOC.isin(no_medal_countries).sort_values(ascending = True)], x = 'Discipline', order = athletes[athletes.NOC.isin(no_medal_countries)].Discipline.value_counts().index)
plt.xticks(rotation = 90)
plt.ylabel('count athletes')
plt.xlabel('Sport')
plt.title('Sports that Countries with No Medals Participated in Olympics 2021')
```

```python
high_medal_countries = c_info[c_info['Total']>= 60].NOC
plt.figure(figsize = (15,5))
sns.countplot(data = athletes[athletes.NOC.isin(high_medal_countries)].sort_values(by = 'Discipline', ascending = True), x = 'Discipline')
plt.xticks(rotation = 90)
plt.xlabel('Sport')
plt.ylabel('count athletes')
plt.title('Sports that Countries with Greater than 60 Medals Participated in Olympics 2021')
```

```python
c_info['medals_per_sport'] = c_info['Total']/c_info['Discipline']
c_info['medals_per_athlete'] = c_info['Total']/c_info['Name']

plt.figure(figsize = (15,5))
sns.barplot(data = c_info.sort_values(by = 'medals_per_sport',ascending = False).head(20), x = 'NOC',y = 'medals_per_sport')
plt.xticks(rotation = 90)
plt.xlabel('Country')
plt.title('Most Successful Countries by Medals per Sport')
```

```python
plt.figure(figsize = (15,5))
sns.barplot(data = c_info.sort_values(by = 'medals_per_athlete',ascending = False).head(20), x = 'NOC',y = 'medals_per_athlete')
plt.xticks(rotation = 90)
plt.xlabel('Country')
plt.title('Most Successful Countries by Medals Per Athlete')
```
